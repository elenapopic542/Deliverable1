! The above three dataframes have a long shape, and there is more than one observation/row for each unit of analysis (country). In the following, for ease of analysis I will transform the datasets into a wide shape before formatting or integrating the data.

*Note: The reshaping can also be done after concatenation. In this, one approach would be to append the dataframes, given that in the long shape all three of my datasets have columns with an identical name, location and datatype. However, for this assignment, I will use the more complex fuzzy merge operations. So I will first reshape all three dataframes into wide format, and then merge them two by two.
 

```python
#Transforming the datasets into a wide format so that each development indicator in the Series column is separated into its own column,
#this ensures there is only one observation per country:

wide_DF1 = DF1csv.pivot(index='Country', columns='Series', values='YR2021')
wide_DF1
```

```python
#Ensuring that the Country is the column's name, and not an index as seen in the above output:

wide_DF1= wide_DF1.reset_index(drop=False).\
             rename_axis(index=None, columns=None)

wide_DF1
```

```python
#Repeating the above operations for the second dataframe:
wide_DF2 = DF2csv.pivot(index='Country', columns='Series', values='YR2021')
wide_DF2

wide_DF2= wide_DF2.reset_index(drop=False).\
             rename_axis(index=None, columns=None)

wide_DF2
```

```python
#Repeating the above operations for the third dataframe:
wide_DF3 = DF3csv.pivot(index='Country', columns='Series', values='YR2021')
wide_DF3

wide_DF3= wide_DF3.reset_index(drop=False).\
             rename_axis(index=None, columns=None)

wide_DF3
```

In the following lines, I will proceed to merging the dataframes two by two. For this operation, one column in each dataframe should have the same unique values (in this case, the Country column). 

```python
#Examining the shape of the first two dataframes:
wide_DF1.shape,wide_DF2.shape
```

Based on the above output, the optimal scenario is for the merged dataframe to have 127 rows. 

```python
#Merging the two dataframes:
wide_DF1.merge(wide_DF2).shape
```

```python
#The merged dataframe has exactly 127 rows, as expected, which means that all the country names were 
#exactly the same in the two dataframes. I save the result as merge1 dataframe:

merge1=wide_DF1.merge(wide_DF2)
```

```python
#Verifying:
merge1.info()
```

```python
#Given that all countries were correctly merged, I proceed to the second merge operation so as to integrate the last dataframe.
#I start again by examining the shape of the relevant dataframes:

merge1.shape,wide_DF3.shape
```

```python
#Based on the above output, the optimal scenario is for the merged dataframe to have 127 rows.

merge1.merge(wide_DF3).shape
```

```python
#Not all countries were matched during the merge. In the following, I identify these countries:

Onlymerge1=set(merge1.Country)-set(wide_DF3.Country)
Onlymerge1
```

```python
#Not all countries were matched during the merge. In the following, I identify these countries:

Onlywide_DF3=set(wide_DF3.Country)-set(merge1.Country)
Onlywide_DF3
```

```python
#Using the fuzzy merge approach to identify matches between Onlywide_DF3 and Onlymerge1.

from thefuzz import process as fz
[(fz.extractOne(merge1, Onlywide_DF3), merge1) for merge1 in sorted(Onlymerge1)]
```

```python
#The above results did not identify any good matches. 
#To confirm, I'm checking the countries with the highest similarity score:

[(fz.extractOne(merge1, Onlywide_DF3),merge1) 
 for merge1 in sorted(Onlymerge1) 
 if fz.extractOne(merge1, Onlywide_DF3)[1]>=86] #86 is the highest similarity score received above
```

```python
#The above result is not a good match. 
#Saving the final merged dataframe:
merge2=merge1.merge(wide_DF3)
```

```python
#Verifying the data types:
merge2.info()
```

```python
#Checking the summary statistics:
merge2.describe()
```

```python
#Creating a boxplot to check for value ranges:
merge2.plot(kind='box', rot=90)
```

Based on the above boxplot, we observe that not all variables in the dataframe are in the same range. 

In the following lines of code, I uniformize the range of values for all the variables in the dataframe using the sklearn package.

```python
import sklearn 
from sklearn import preprocessing
    
#Creating a MinMaxScaler object based on the desired range of values
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))

# Since multiple variables in the dataframe are out of the desired range, I will not create arrays for each variable, 
# but rather create an array of all columns containing numeric variables in the merge2 dataset using the above scaler
numeric_cols = merge2.columns[1:10]
array = scaler.fit_transform(merge2[numeric_cols])

print(array.shape)
```

```python
#Checking the datatype
type(array)
```

```python
#Replacing the values of the numeric variables with the above defined array:
merge2[numeric_cols] = array
```

```python
#Re-creating the boxplot:
merge2.plot(kind='box', rot=90)
```

Having finished the data cleaning, formatting and merging operations, we proceed to exporting the file into an R dataframe. 

```python
#Saving the Pandas dataframe as an RDS file using the rpy2 package:
import rpy2
from rpy2.robjects import pandas2ri
pandas2ri.activate()

from rpy2.robjects.packages import importr

base = importr('base')
base.saveRDS(merge2,file="wbdata.RDS")
```

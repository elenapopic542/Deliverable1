In the following lines, we will check whether the conditions for concatenating the dataframes are met, namely: 
all three dataframes should have columns with an identical name, location and datatype.

```python
# Examining the columns that are shared among the three dataframes:
set(DF1csv.columns)&set(DF2csv)&set(DF3csv)
```

```python
#Keeping only the columns that are shared among the three datasets:
common=set(DF1csv.columns)&set(DF2csv)&set(DF3csv)

# we are keeping just the 'common' columns:
DF1csvap=DF1csv.loc[:,common]
DF2csvap=DF2csv.loc[:,common]
DF3csvap=DF3csv.loc[:,common]
```

```python
#Verifying if the data in the same columns have an identical datatype:
theZips=zip(DF1csv.dtypes,
            DF2csv.dtypes,
            DF3csv.dtypes)

[x for x in theZips if len(set(x))>1]
```

```python
#Given that all three dataframes now have columns with an identical name, location and datatype, 
#we will integrate the dataframes by concatenating:
Inflationcsv=pd.concat([DF1csv,DF2csv,DF3csv])
Inflationcsv
```

In the above dataset, the countries are repeated three times for each of the three values of the indicator type: GDPgrowth, GNI and inflation. In the following section, we will reshape the dataset from long to wide format.  

```python
InflationWide = pd.pivot_table(Inflationcsv,
               values=['2021'], 
               index=['CountryName'],
               columns=['IndicatorName']) 
InflationWide
```

```python
InflationWide = InflationWide.reset_index(drop=False).\
             rename_axis(index=None, columns=None)
```

```python
#Creating a new year column
newcolumn = (2021)
 
InflationWide = InflationWide.assign(Year=newcolumn)

InflationWide
```
